### Questão 2
Aplique o k-means na base de dados de comparação de produtos da ShopMania. Analise seus resultados. A seguir, o link para o dataset:

https://www.kaggle.com/datasets/lakritidis/product-classification-and-categorization

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from scipy.sparse import hstack

# Carregar os dados
df = pd.read_csv("shopmania.csv")

# Nomear cada coluna
df.columns = ["id", "product", "category_id", "category"]

# Transformar os produtos e categorias em features numéricas para utilizá-los no k-means
tfidf = TfidfVectorizer()
feature_product = tfidf.fit_transform(df["product"])
ohe = OneHotEncoder()
feature_category = ohe.fit_transform(df[["category"]])
features = hstack([feature_product, feature_category])

# O contexto dos dados ShopMania diz que os dois níveis mais altos da hierarquia incluem 
#   39 categorias. Este será então o ponto de partida para escolher o valor de k.
N_CLUSTERS = 39
RANDOM_STATE = 11

# Criar o modelo com o uso do algorítmo k-means
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE)
assignments = kmeans.fit_predict(features)

# Calcular a inercia
score = kmeans.inertia_
print(f"Inércia: {score}")

# Calcular número de itens dentro de cada cluster
import numpy as np
counts = np.bincount(assignments)
for idx, count in enumerate(counts):
    print(f"Cluster {idx}: {count} itens")
```

### Questão 5
O dataset da ShopMania (questão 2) inclui 230 categorias. Use o PCA para reduzir a dimensionalidade, refaça o processo de agrupamento e compare os resultados com aquele obtido na questão 2.


```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from scipy.sparse import hstack, csr_matrix
from sklearn.decomposition import PCA

# Carregar os dados
df = pd.read_csv("shopmania.csv")

# Nomear cada coluna
df.columns = ["id", "product", "category_id", "category"]

# Transformar os produtos e categorias em features numéricas para utilizá-los no k-means
tfidf = TfidfVectorizer()
feature_product = tfidf.fit_transform(df["product"])
ohe = OneHotEncoder()
feature_category = ohe.fit_transform(df[["category"]])
scaler = StandardScaler(with_mean=False)
feature_category_scaled = scaler.fit_transform(feature_category).toarray()

# Aplicar PCA nas categorias
pca = PCA(n_components=2)
feature_category_pca = pca.fit_transform(feature_category_scaled)
features = hstack([feature_product, csr_matrix(feature_category_pca)])

# O contexto dos dados ShopMania diz que os dois níveis mais altos da hierarquia incluem 
#   39 categorias. Este será então o ponto de partida para escolher o valor de k.
N_CLUSTERS = 39
RANDOM_STATE = 11

# Criar o modelo com o uso do algorítmo k-means
kmeans = KMeans(n_clusters=N_CLUSTERS, random_state=RANDOM_STATE)
assignments = kmeans.fit_predict(features)

# Calcular a inercia
score = kmeans.inertia_
print(f"Inércia: {score}")

# Calcular número de itens dentro de cada cluster
import numpy as np
counts = np.bincount(assignments)
for idx, count in enumerate(counts):
    print(f"Cluster {idx}: {count} items")
```
